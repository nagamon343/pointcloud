```python
import numpy as np

from pyntcloud import PyntCloud

import pyvista as pv
```

```python
anky = PyntCloud.from_file("data/ankylosaurus_mesh.ply")

converted_triangle_mesh = anky.to_instance("pyvista", mesh=True)
```

```python
# 一回scalars='RGB'ってしないとだめ？
converted_triangle_mesh.plot(scalars='RGB', rgb=True)
```

```python
converted_triangle_mesh.plot(rgb=True, use_panel=True)
```

```python
plotter = pv.Plotter()

#plotter.add_mesh(pppline)
plotter.add_mesh(converted_triangle_mesh, rgb=True)

# 遠近法じゃないモードになる
plotter.enable_parallel_projection()

# これ　カメラの向き
plotter.view_isometric()
plotter.show()
```
